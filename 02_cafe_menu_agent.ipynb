```python
import os
from google import genai
from google.genai import types
```

```python
# 1) Твої tool-функції (як в Azure прикладі)
def get_specials() -> str:
    """Returns today's specials."""
    return (
        "Special Soup: Clam Chowder\n"
        "Special Salad: Cobb Salad\n"
        "Special Drink: Chai Tea\n"
    )
```

```python
def get_item_price(menu_item: str) -> str:
    """Returns price for the menu item."""
    # demo-логіка
    prices = {
        "Clam Chowder": "$8.50",
        "Cobb Salad": "$10.00",
        "Chai Tea": "$3.50",
    }
    return prices.get(menu_item, "$9.99")
```

```python
# 2) Клієнт Gemini (ключ або з env, або передай api_key=...)
# export GEMINI_API_KEY="YOUR_KEY"
client = genai.Client(api_key=os.environ.get("GEMINI_API_KEY"))
```

```python
config = types.GenerateContentConfig(
    tools=[get_specials, get_item_price],
    # за потреби можна керувати AFC (наприклад, вимкнути):
    automatic_function_calling=types.AutomaticFunctionCallingConfig(disable=False),
)
```

```python
chat = client.chats.create(
    model="gemini-3-flash-preview",
    # model="gemini-2.5-pro",
    config=config,
)
```

```python
user_inputs = [
    # "Where is Berlin?",
    "Hello",
    "What is the special soup?",
    "How much does that cost?",
    "Thank you",
]

for user_input in user_inputs:
    print(f"# User: {user_input}")
    resp = chat.send_message(user_input)
    print(f"# Model: {resp.text}\n")
```
