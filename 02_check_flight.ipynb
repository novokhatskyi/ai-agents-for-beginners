```python
import os
from google import genai
from google.genai import types
```

```python
def check_destination_availability(destination: str) -> str:
    available = {
        "Barcelona": True,
        "Tokyo": True,
        "Cape Town": False,
        "Vancouver": True,
        "Dubai": False,
    }
    is_available = available.get(destination, False)
    return f"{destination} is {'available' if is_available else 'not available'}."
```

```python
client = genai.Client(api_key=os.environ["GEMINI_API_KEY"])

config = types.GenerateContentConfig(
    system_instruction=(
        "You are a travel booking agent. "
        "Always check availability before recommending a destination. "
        "Use the tool check_destination_availability(destination)."
    ),
    tools=[check_destination_availability],
)

chat = client.chats.create(model="gemini-3-flash-preview", config=config)

print(chat.send_message("Which destinations do you have available?").text)
print(chat.send_message("I'd like to go somewhere warm. What's available?").text)
```
